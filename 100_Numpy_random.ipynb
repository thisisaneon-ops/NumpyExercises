```python
import numpy as np
x = np.arange(1,11)
print(x)
i = 0; j = 9
while i < j:
    tmp = x[i]
    x[i] = x[j]
    x[j] = tmp
    i = i + 1
    j = j - 1
print(x)
y = np.arange(1,11)
print(y)
y_reversed = y[::-1]
print(y_reversed)
```
